# Importing all the important modules for the project

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
import sklearn
from sklearn.tree import plot_tree
import warnings

%matplotlib inline
sns.set_theme()
```

# Creating the DataFrame

```python
data = pd.read_csv('autoscout24.csv')
```

# Renaming the columns of the DataFrame and dropping NaN
# Removing every value for 'Jahr' that is smaller/equal 2011

```python
data = data.rename(columns={'mileage': 'Kilometerstand', 'make': 'Hersteller', 'model': 'Modell', 'fuel': 'Kraftstoff', 'gear': 'Schaltung', 'offerType': 'Zustand', 'price': 'Preis', 'hp': 'PS', 'year': 'Jahr'})
data = data.dropna()
data = data[data['Jahr'] >= 2011]
```

# Using pd.options.display.float_format = '{:.0f}'.format to change the numerical Values shown in the DataFrame to format float numbers as integers 

```python
pd.options.display.float_format = '{:.0f}'.format
```

# Using data.describe() to show the numerical value data and using indexing/slicing to change from scientific format to integers

```python
data['Kilometerstand'] = data['Kilometerstand'].round(0).astype(int)
data['Preis'] = data['Preis'].round(0).astype(int)
data['PS'] = data['PS'].round(0).astype(int)
data['Jahr'] = data['Jahr'].round(0).astype(int)
data.describe()
```

# 1.3.1 Top 5 meistverkaufte Hersteller im Datensatz

## Filtering and Creating a new DataFrame for the top 5 most sold manufacturers using value_counts(), isin() and .index()
## Using for loop to print the top 5 manufactureres

```python
hersteller_count = data['Hersteller'].value_counts()
data_five = data[data['Hersteller'].isin(hersteller_count.index[:5])]
for i in hersteller_count.index[:5]:
    print(i)
```

```python
data_five.describe()
```

# 1.3.2 Wie teuer war durchschnittlich ein Auto von Hersteller X

## Using groupby() function to group the Data by 'Hersteller', Filtering the 'Preis', using mean() to get the average 'Preis' of the 'Hersteller' and then sorting it by highest Value using sort_values(ascending=False)

```python
data_five.groupby(['Hersteller']).Preis.mean().sort_values(ascending=False)
```

```python
sns.histplot(data=data_five.groupby(['Hersteller']).count().sort_values(ascending=False, by='Preis'))
```

# 1.3.3 Trainiere ein Modell deiner Wahl, z.b: LinReg, um ausgehend von Features deiner Wahl den Verkaufspreis eines Autos vorherzusagen.

## creating the variable y with the label 'Preis' and variable X with our features ['Kilometerstand', 'PS', 'Jahr']
## using train_test_split to split our datasets into train and test data
## using LinearRegression() and model.fit(X_train, y_train) on our training data to fit our trained model
## using model.predict(X_test) on our test data to create predictions

```python
y = data_five['Preis']
X = data_five[['Kilometerstand', 'PS', 'Jahr']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

## plotting our Predictions and Correct Labels to show the Model

```python
plt.scatter(y_test, predictions)
plt.xlabel("Korrekte labels: y_test")
plt.ylabel("Predictions")
```

```python
koeffizient = pd.DataFrame(model.coef_, X.columns)
koeffizient.columns = ['Umsatz']
koeffizient
```

# 1.3.4 Welche Kategorie von machine Learning ist dieses Problem 

Lineare Regression ist Teil des 'supervised machine learning'. Dabei lernt der Algorithmus wie er Input Daten auf ein bestimmtest Ziel zuweist. Lineare Regression wird zur Vorhersage von Kontinuierlichen Zielen verwendet indem man die Beziehung zwischen den inputs und den targets als lineare Gleichung darstellt. Oftmals wird das Modell genutzt um Regressions Analyse und Predictive Modelling durchzufÃ¼hren wo das Ziel ist basierend auf dem Input eine Vorhersage zu erstellen. Ein Beispiel hierfÃ¼r ist die Marktanalyse in der Wirtschaft.

# 1.3.5 Wie wird die GÃ¼te des Modells evlauiert? Was ist die Fehlermatrik?

Um die GÃ¼te des Modells zu evaluieren Verwenden wir verschiedene Fehlermetriken. In diesem Projekt schauen wir uns besonders Modelle die die Fehler der Regression darstellen an. DafÃ¼r wird hier der Mean Absolute Error (MAE) und der Mean Squared Error (MSE) verwendet. Meist wird der Root Squared Error (RSE) verwendet.

```python
print("MAE", metrics.mean_absolute_error(y_test, predictions))
print("MSE", metrics.mean_squared_error(y_test, predictions))
print("RSE", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

Der MAE misst die durchschnittliche Abweichung von den Fehlern. Je kleiner der MAE desto genauer ist das Modell. Problem = Ausnahmen

Je hÃ¶her die Zahlen desto mehr AusreiÃŸer hat das Modell.

```python
sns.histplot((y_test - predictions), kde=True)
```

Man sieht, dass sich die Daten an einer Normalverteilung nÃ¤hern sodass das Ergebnis "gut" ist.
